# Time Plot

> The timePlot function is designed to quickly plot time series of data, perhaps for
several pollutants or variables. This is, or should be, a very common task in the analysis
of air pollution. In doing so, it is helpful to be able to plot several pollutants at the same
time (and maybe other variables) and quickly choose the time periods of interest. It
will plot time series of type Date and hourly and high time resolution data.

Standard libraries to be imported for usage

```python
import pandas as pd
import matplotlib.pyplot as plt

from vayu import timePlot
```

Reading the data and using the function from the library.

```python
Reading the data and using the function from the library.mydata = pd.read_csv('mydata.csv')
print(timePlot(mydata,'2003', 8))
```
